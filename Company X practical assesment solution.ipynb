```r
#Load libraries, Set working directory, Import files
library(readxl)

setwd("C:/Users/Steve/Desktop/Portfolio/X Interview - Case Study - Data Officer")

raw_roster <- read_xlsx("Final Case Study - Data Officer- September 2018 - Copy.xlsx", sheet = "Academy Staff Rooster")
raw_jtitles <- read_xlsx("Final Case Study - Data Officer- September 2018 - Copy.xlsx", sheet = "Job Titles")
```

```r
#Question 1
library(dplyr)
library(stringr)

q1 <- raw_jtitles %>%
  mutate(nn = if_else(Department == "Executive" | str_detect(`Rank/Level`, "Chief"), str_c(str_replace_na(Specialisation), " ", `Rank/Level`),
                      if_else(str_detect(`Rank/Level`, "Lead"), str_c("Lead ", str_replace_na(Specialisation), " ", str_remove(`Rank/Level`, "Lead"), ", ", Department ),
                              if_else(str_detect(`Rank/Level`, "Senior"), str_c("Senior ", str_replace_na(Specialisation), " ", str_remove(`Rank/Level`, "Senior"), ", ", Department ),
                                      if_else(str_detect(Department, "Leadership") | str_detect(raw_jtitles$Department, "Leadership") & is.na(raw_jtitles$Specialisation), str_c(str_replace_na(Specialisation), " ", `Rank/Level`, ", ", Group), str_c(str_replace_na(Specialisation), " ", `Rank/Level`, ", ", Department))))))

q1$`Updated Job title` <- str_remove(q1$nn, "NA ")

q1 <- select(q1, - nn)

head(q1)
```

```r
#Question 2
q2 <- q1 %>% group_by(`Updated Job title`) %>%
  count(`Updated Job title`) %>%
  arrange(desc(n))

q2
```

```r
#Question 3
q3 <- q1 %>% group_by(Territory) %>%
  count(Territory) %>%
  arrange(desc(n))

q3
```

```r
#Question 4 - i
q4 <- raw_roster %>%
  group_by(Status) %>% 
  count(`Job Position`) %>%
  arrange(desc(n))

q4i <- filter(q4, Status == "Active")

head(q4i)
```

```r
#Question 4 - ii
raw_roster <- raw_roster %>%
  mutate(age = 2020 - as.numeric(str_sub(`Date of Birth`,1,4)))

head(raw_roster)
```

```r
#Question 4 - iii
q4iii <- raw_roster %>%
  group_by(Gender) %>%
  summarise(median(age, na.rm = TRUE))

q4iii
```

```r
#Question 4 - iv
q4iv <- aggregate(raw_roster[, 18], list(raw_roster$`Base Academy`), mean) %>%
  filter(age %in% range(age))

q4iv
```

```r
#Question 4 - v
#Turnover at base academies
q4va <- raw_roster %>% group_by(`Base Academy`, Status) %>%
  count(Status) %>%
  arrange(desc(n))

q4va
```

```r
#Base Academies with Active Teachers who are Degree holders
q4vb <- raw_roster %>%
  group_by(`Base Academy`, Status) %>%
  count(`Highest Level Education`) %>%
  filter(`Highest Level Education` == "Degree", Status == "Active")

q4vb
```
